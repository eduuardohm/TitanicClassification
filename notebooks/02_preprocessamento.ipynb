```python
import pandas as pd
from pathlib import Path
```

```python
BASE_PATH = Path("../data")

RAW_PATH = BASE_PATH / "raw"
PREPROCESSED_PATH = BASE_PATH / "preprocessed"

# cria a pasta caso ela não exista
PREPROCESSED_PATH.mkdir(parents=True, exist_ok=True)
```

```python
train_df = pd.read_csv(RAW_PATH / "train.csv")
test_df = pd.read_csv(RAW_PATH / "test.csv")
```

```python
def preprocess_data(df):
    
    columns_to_remove = [
        "PassengerId",
        "Name",
        "Ticket",
        # "Fare",
        "Cabin",
    ]
    df = df.drop(columns=columns_to_remove)
    
    mean_age_female = df[df["Sex"] == "female"]["Age"].mean()
    mean_age_male = df[df["Sex"] == "male"]["Age"].mean()
    
    female_condition = (
        (df["Sex"] == "female") &
        (df["Age"].isna())
    )
    male_condition = (
        (df["Sex"] == "male") &
        (df["Age"].isna())
    )
    
    df.loc[female_condition, "Age"] = mean_age_female
    df.loc[male_condition, "Age"] = mean_age_male

    df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
    df = df.drop(columns=["SibSp", "Parch"])
    
    return df
```

```python
train_processed = preprocess_data(train_df)
test_processed = preprocess_data(test_df)
```

```python
train_processed.to_csv(
    PREPROCESSED_PATH / "train_preprocessed.csv",
    index=False
)

test_processed.to_csv(
    PREPROCESSED_PATH / "test_preprocessed.csv",
    index=False
)
```
